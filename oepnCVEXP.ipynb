```python
#   引入依赖
import cv2
import matplotlib.pyplot as plt

#   打开视频
video = cv2.VideoCapture("./datasets/BVN.mp4")
print(video)
#   返回flag为是否成功读取 若未读取结束或未成功读取则返回false frame是一帧的内容
flag, frame = video.read()
#   展示图片
# plt.imshow(frame)
#   默认生成的为BGR格式 要转换为RGB 颜色才正确
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
```

```python
video = cv2.VideoCapture("./datasets/BVN.mp4")
num = 0
save_step = 30
while True:
    flag,frame = video.read()

    if not flag:
        break

    num += 1

    if num % save_step == 0:
        cv2.imwrite(f"./datasets/images/{str(num)}.jpg",frame)
```
