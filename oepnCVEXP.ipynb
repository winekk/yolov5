```python
video = cv2.VideoCapture("./datasets/BVN.mp4")
num = 0
save_step = 30
while True:
    flag,frame = video.read()

    if not flag:
        break

    num += 1

    if num % save_step == 0:
        cv2.imwrite(f"./datasets/images/{str(num)}.jpg",frame)
```
